```python
#이진탐색트리 클래스
class BinarySearchTree:
  class Node:
    def __init__(self, data= None, leftChild=None, rightChild=None):
      self.data = data
      self.left = leftChild
      self.right = rightChild
  
  def __init__(self):
    self.root = None
  #-전위순회---------------------------
  def preShow(self):
    print('전위순회:', end = '')
    self.__preOrder(self.root)
    print()

  def __preOrder(self, node): # 전위순회 
    if node != None:
      print(node.data, end = ', ')
      self.__preOrder(node.left)
      self.__preOrder(node.right)
  #-------------------------------------
  #-중위순회----------------------------
  def inShow(self):
    print('중위순회 : ', end = '')
    self.__inOrder(self.root)
    print()


  def __inOrder(self, node):
    if node != None:
      self.__inOrder(node.left)
      print(node.data, end=', ')
      self.__inOrder(node.right)
  #-------------------------------------
  #-후위순회----------------------------
  def postShow(self):
    # 함수 내부에 함수를 정의했기 때문에
    # postOrder() 함수는 오로지 postShow()메소드 내에서만 사용이
    # 가능하다
    def postOrder(node):
      if node != None:
        postOrder(node.left)
        postOrder(node.right)
        print(node.data, end = ', ')

    print('후위순회 : ', end = '')
    postOrder(self.root)
    print()
  #------------------------------------




  #추가
  def add(self, data):
    newNode = self.Node(data)

    #트리가 비어있다면
    if self.root == None:
      self.root = newNode
      return
    
    curr = self.root
    while True:
      # 삽입하고자 하는 데이터가 노드보다 크다면
      if curr.data < data:
        if curr.right == None: # 만약 오른쪽 자식이 비었다면
          curr.right = newNode # 해당 노드의 오른쪽 자식으로 추가
          break
        #오른쪽에서 계속 탐색
        curr = curr.right
      #삽입하고자 하는 데이터가 노드보다 작다면
      elif curr.data > data:
        if curr.left == None: # 왼쪽자식이 비었다면 
          curr.left = newNode # 왼쪽 자식으로 새로만든 노드 추가
          break
        #왼쪽에서 계속 탐색
        curr = curr.left
      else : # curr.data == data 중복되는 데이터가 있다면
        return # 추가 실패

  #삭제
  def remove(self, key):
    # 만약 비어있다면 : 실패
    if self.root == None : return

    parent = self.root # 삭제할 노드의 부모를 담아줄 변수
    target = self.root # 삭제할 노드를 담아줄 변수

    while target != None:
      if target.data == key: break
      else:
        parent = target
        if key > target.data:
          target = target.right
        else: # key < target.data
          target = target.left
    
    # 1. while문이 끝나고 나면 target에는 내가 삭제할 노드
    # parent에는 삭제할 노드의 부모 노드가 담겨있다
    # 2. 우리가 삭제하고자 하는 요소가 트리에 들어있지 않다면
    # target에는 None이 들어있다
    # 3. 만약 삭제하고자 하는 요소가 루트라면 target과 parent에는 
    # root 노드가 들어있다

    #삭제할노드를찾지못했다면 : 실패
    if target == None: return

    #삭제할노드가 리프노드(자식이 없다)라면
    if target.right == None and target.left == None:
      if parent == target:
        self.root = None
      elif parent.left == target: #key > parent.data
        #삭제할 노드가 왼쪽 자식이라면
        parent.left = None
      else : #parent.right == target
        parent.right = None

    #삭제할노드가 두 자식을 갖고 있다면
    elif target.right and target.left:
      #계승할 노드 찾기
      #왼쪽 서브트리에서 가장 큰수 / 오른쪽 서브트리에서 가장 작은수
      succ_parent = target # 계승할 노드의 부모노드
      succ = target.left # 계승할 노드

      while succ.right != None:
        succ_parent = succ
        succ = succ.right
      
      # 반복문이 끝나고 나면 succ에는 계승할 노드, succ_parent에는
      # 계승할 노드의 부모노드가 들어있다
      target.data = succ.data
      if succ_parent.right == succ:
        succ_parent.right = succ.left
      else: #succ_parent.left == succ:
        succ_parent.left = succ.left


    #삭제할노드가 한 자식을 갖고 있다면
    else : 
      # child: 삭제할 노드의 자식을 담아줄 변수
      # 만일 오른쪽자식이 있다면(오른쪽자식이 None이 아니라면)
      # child에는 오른쪽 자식이 담기고
      # 그거 아니면 child에는 왼쪽 자식이 담긴다
      child = target.right or target.left

      if target == parent:# 삭제하고자 하는 노드가 루트라면
        self.root = child
      elif target == parent.left: #삭제할 노드가 왼쪽 자식이라면
        parent.left = child
      else: # 삭제할 노드가 오른쪽 자식이라면
        parent.right = child



  #검색
  def select(self, key):
    curr = self.root
    while curr!=None:
      if key == curr.data: break
      elif key > curr.data: curr = curr.right
      else : curr = curr.left
    
    #찾기 실패의 경우 curr에는 None이 들어있고
    #찾은 경우 curr에는 찾고자 하는 노드가 들어있다

    # 만약 요소가 없다면 : 실패
    if curr == None: return
    else : return curr.data

  def re_select(self, key):
    findNode = self.re_select_value(self.root, key)
    if findNode != None:
      return findNode.data

  # 재귀함수를 통한 검색
  def re_select_value(self, node, key):
    if node == None or node.data == key:
      return node
    
    if key < node.data:
      return self.re_select_value(node.left, key)
    
    if key > node.data:
      return self.re_select_value(node.right, key)


  # 재귀함수를 통한 값의 추가
  def re_add(self, data):
    self.root = self.re_add_value(self.root, data)
    
  
  def re_add_value(self, node, data):
    if node == None:
      node = self.Node(data)
    else:
      if data == node.data: return
      elif data < node.data:
        node.left = self.re_add_value(node.left, data)
      else:
        node.right = self.re_add_value(node.right, data)

    return node


  def re_remove(self, data):
    self.root, removeNode = self.re_remove_value(self.root, data)
    removeNode.left = None
    removeNode.right = None

  #재귀함수 삭제
  def re_remove_value(self, node, data):
    if node != None: # 삭제할 요소가 트리에 없다면
      return None, None
    
    #삭제하고자 하는 데이터가 노드의 데이터보다 작다면
    elif data < node.data:
      node.left, removeNode = self.re_remove_value(node.left, data)
    #삭제하고자 하는 데이터가 노드의 데이터보다 크다면
    elif data > node.data:
      node.right, removeNode = self.re_remove_value(node.right, data)
    else: #내가 삭제하고자 하는 노드를 찾은경우
      # 삭제하고자 하는 노드의 자식이 없는 경우
      if node.left == None and node.right == None:
        removeNode = node
        node = None
      # 삭제하고자 하는 노드가 두 자식을 갖는 경우
      elif node.left and node.right:
        #계승할 노드
        succ = node.left
        while succ.right != None:
          succ = succ.right
        
        node.data, succ.data = succ.data, node.data
        node.left, removeNode = self.re_remove_value(node.left, succ.data)

      # 자식노드가 하나일때
      else:
        removeNode = node
        node = node.left or node.right
      
    return node, removeNode
```

```python
tree = BinarySearchTree()
tree.re_add(10)
tree.re_add(20)
tree.re_add(20)
tree.re_add(15)
tree.inShow()
```

```python
tree = BinarySearchTree()

tree.add(10)
tree.add(5)
tree.add(15)
tree.add(3)
tree.add(4)
tree.add(20)
tree.add(12)
tree.preShow()
tree.inShow()
tree.postShow()

tree.remove(20)
tree.inShow()
tree.remove(3)
tree.inShow()
tree.remove(15)
tree.inShow()
tree.remove(10)
tree.inShow()

print('찾은 값:', tree.select(500))
```
